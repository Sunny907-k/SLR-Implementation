```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
#reading the data
#here we are loading the data from scikit learn

from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes
```

```python
#Problem statement : To predict disease progression one year adter baseline provided 10 features
#In one year, how the sugar level has changed
```

```python
print(diabetes.DESCR)
```

```python
diabetes.feature_names
```

```python
data = pd.DataFrame(diabetes.data, columns = diabetes.feature_names)
```

```python
data
```

```python
data['target'] = diabetes.target
```

```python
data
```

```python
#EDA, Data cleaning, data preperation, feature engineering
```

```python
X = data[["bmi"]] # one independant variable
y = data["target"] # dependant variable
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=1)
#Random_state when fixed helps us to reproduce the same result again and again
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#scaling(optional)
```

```python
#model training

from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
X_train
```

```python
y_train
```

```python
#training of model
model.fit(X_train, y_train)
#at the backend, it will use gradient descent to give optinal coefficient
```

```python
model.coef_ #attribute of the model
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#visualise the result

plt.scatter(X_test, y_test, color = "black", label = "Actual" )
plt.plot(X_test, y_pred, color = "blue", linewidth = 3, label = "Prediction Linear Regression")
plt.xlabel("BMI")
plt.ylabel("One Year Progression Target")
plt.title("Linear Regression on Diabetes data")
plt.legend()
plt.show()
```

```python
#model evaluation

from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```
